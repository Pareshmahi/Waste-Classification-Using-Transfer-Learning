```python
model_accuracies = {
    "ResNet-50": 0.8977,
    "ResNet-34": 0.8922,
    "DenseNet-121": 0.9162,
    "EfficientNet-B0": 0.9261,
    "EfficientNet-B3": 0.9464,
    "Inception-V3": 0.9162,
    "VGG16": 0.8860,
    "MobileNet-V2": 0.8983
}
```

```python
import matplotlib.pyplot as plt

# Define colors for each model
colors = ["b", "g", "r", "c", "m", "y", "k", "#FF5733"]

# Extract model names and accuracies
models = list(model_accuracies.keys())
accuracies = list(model_accuracies.values())

# Convert accuracy to percentage
accuracies_percent = [acc * 100 for acc in accuracies]

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(models, accuracies_percent, color=colors, alpha=0.7)

# Add labels
plt.xlabel("Model Type")
plt.ylabel("Validation Accuracy (%)")
plt.title("Model Type vs. Validation Accuracy")
plt.xticks(rotation=25)
plt.ylim(80, 100)  # Set Y-axis range for better visualization
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Show values on bars
for i, acc in enumerate(accuracies_percent):
    plt.text(i, acc + 0.5, f"{acc:.1f}%", ha="center", fontsize=10, fontweight="bold")

plt.show()
```
