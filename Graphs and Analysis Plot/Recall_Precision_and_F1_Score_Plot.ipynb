```python
model_metrics = {
    "ResNet-50": {"Precision": 0.8997, "Recall": 0.8957, "F1-score": 0.8971},
    "ResNet-34": {"Precision": 0.8934, "Recall": 0.8861, "F1-score": 0.8889},
    "DenseNet-121": {"Precision": 0.9210, "Recall": 0.9138, "F1-score": 0.9159},
    "EfficientNet-B0": {"Precision": 0.9305, "Recall": 0.9222, "F1-score": 0.9255},
    "EfficientNet-B3": {"Precision": 0.9466, "Recall": 0.9463, "F1-score": 0.9462},
    "Inception-V3": {"Precision": 0.9187, "Recall": 0.9129, "F1-score": 0.9151},
    "VGG16": {"Precision": 0.8863, "Recall": 0.8845, "F1-score": 0.8849},
    "MobileNet-V2": {"Precision": 0.9034, "Recall": 0.8950, "F1-score": 0.8979}
}
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extract data for plotting
models = list(model_metrics.keys())
precision = [model_metrics[m]["Precision"] for m in models]
recall = [model_metrics[m]["Recall"] for m in models]
f1_score = [model_metrics[m]["F1-score"] for m in models]

x = np.arange(len(models))  # Label locations
width = 0.2  # Bar width

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot bars for each metric
rects1 = ax.bar(x - width, precision, width, label="Precision", color="b", alpha=0.7)
rects2 = ax.bar(x, recall, width, label="Recall", color="g", alpha=0.7)
rects3 = ax.bar(x + width, f1_score, width, label="F1-score", color="r", alpha=0.7)

# Add labels, title, and legend
ax.set_xlabel("Model Type")
ax.set_ylabel("Score")
ax.set_title("Precision, Recall, and F1-score per Model")
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=25)
ax.legend()
ax.set_ylim(0, 1)  # Set Y-axis range

# Show values inside bars (90-degree rotated)
for rect in rects1 + rects2 + rects3:
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height - 0.05, f"{height:.2f}",
            ha="center", va="top", fontsize=9, rotation=90, color="white", fontweight="bold")  # Inside the bars

plt.show()
```
