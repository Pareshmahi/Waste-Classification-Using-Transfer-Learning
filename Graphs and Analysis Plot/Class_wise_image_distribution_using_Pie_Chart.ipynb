```python
import os
import matplotlib.pyplot as plt
```

```python
# Path to dataset
dataset_path = "/content/drive/MyDrive/Colab Notebooks/Waste Classification using Transfer Learning/Dataset"

# Get class-wise image count
class_counts = {}
for class_name in os.listdir(dataset_path):
    class_folder = os.path.join(dataset_path, class_name)
    if os.path.isdir(class_folder):  # Ensure it's a folder
        class_counts[class_name] = len(os.listdir(class_folder))

# Print class counts
print("Image count per class:")
for cls, count in class_counts.items():
    print(f"{cls}: {count}")
```

```python
# Plot Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(class_counts.values(), labels=class_counts.keys(), autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6', '#ff6666'])
plt.title("Dataset Class Distribution")
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```
