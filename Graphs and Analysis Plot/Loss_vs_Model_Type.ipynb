```python
model_losses = {
    "ResNet-50": 0.4013,
    "ResNet-34": 0.3832,
    "DenseNet-121": 0.3264,
    "EfficientNet-B0": 0.3129,
    "EfficientNet-B3": 0.2034,
    "InceptionNet-V2": 0.3299,
    "VGG16": 0.4805,
    "MobileNet-V2": 0.4084
}
```

```python
import matplotlib.pyplot as plt

# Define colors for each model
colors = ["b", "g", "r", "c", "m", "y", "k", "#FF5733"]

# Extract model names and losses
models = list(model_losses.keys())
losses = list(model_losses.values())

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(models, losses, color=colors, alpha=0.7)

# Add labels
plt.xlabel("Model Type")
plt.ylabel("Validation Loss")
plt.title("Model Type vs. Validation Loss")
plt.xticks(rotation=25)
plt.ylim(0, max(losses) + 0.05)  # Set Y-axis range for better visualization
plt.grid(axis="y", linestyle="--", alpha=0.5)

# Show values on bars
for i, loss in enumerate(losses):
    plt.text(i, loss + 0.01, f"{loss:.2f}", ha="center", fontsize=10, fontweight="bold")

plt.show()
```
